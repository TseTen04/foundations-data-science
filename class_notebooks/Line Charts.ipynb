# Line Charts and Manipulating the Index

```python
import pandas
```

```python
df = pandas.read_csv('https://bit.ly/music_industry_data')
```

```python
df
```

```python
value_bools = df.metric == 'Value'

value_df = df[value_bools]
```

```python
cd_bools = df.format == 'CD'

cd_df =value_df[cd_bools]
```

```python
cd_df
```

```python
cd_df.index = cd_df.year
```

```python
cd_df.head(2)
```

```python
cd_sales = cd_df.value_actual
```

```python
cd_sales.plot(kind='line', ylabel="Sales in Millions")
```

```python
vinyl_df = value_df[value_df.format == 'LP/EP']
```

```python
vinyl_df.index = vinyl_df.year

vinyl_sales = vinyl_df.value_actual
```

```python
sales_comparison_df = pandas.DataFrame({
        'CD Sales': cd_sales,
    'Vinyl Sales': vinyl_sales,
})
```

```python
sales_comparison_df.plot(kind='line')
```
