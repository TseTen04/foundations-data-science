# Creating Charts

```python
import pandas
```

```python
df = pandas.read_csv('http://bit.ly/airbnbcsv')
```

The neighborhood group is the borough of the listing. Let's try to make a pie chart showing the proportions of the different boroughs. To do that, we'll first count the occurances of the boroughs in the neighborhood_grup column.

```python
df.neighbourhood_group
```

```python
borough_counts = df.neighbourhood_group.value_counts()

borough_counts
```

The value_counts() method on the neighbourhood_group series counts the occurances of each item in the series. It returns a new series of the counts. The index (the labels on the left) are the items that got counted, in this case the boroughs.

The .index attribute lets us see in index of a series (the label part on the left when the series is displayed).

```python
borough_counts.index
```

We can use the counts series we created to generate a pie chart. The index is used to label each slice of the pie. We can also pass in various arguments to change how the pie chart will look, such as changing the color map. The "kind" argument determines broadly what kind of chart will be generated.

```python
borough_counts.plot(kind='pie', ylabel='', title='Borough Percentage', 
                    cmap='magma', figsize=[5,5],
                    fontsize=16)
```

This data also lends itself to a bar chart.

```python
borough_counts.sort_values().plot(kind='bar', xlabel="Borough",
                    ylabel="Number of Listings")
```

It often helps to do various transformations on your data before plotting to get your desired result. Here we have decided to plot the number of listing for each neighborhood, but there are too many results, which looks like a mess:

```python
df.neighbourhood.value_counts().plot(kind='bar')
```

Before plotting, let's use the .head() method to extract only the top 30 neighbourhoods by number of listings, which should be more tidy.

```python
df.neighbourhood.value_counts().head(30).plot(kind='bar')
```
