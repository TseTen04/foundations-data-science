```python
industry = pandas.read_csv('https://bit.ly/music_industry_data')
```

```python
industry

units = industry[industry.metric == 'Units']
```

```python
cd = units[units.format == 'CD']

vinyl = units[units.format == 'LP/EP']
```

```python
cd.index, vinyl.index = cd.year, vinyl.year
```

```python
df = pandas.DataFrame({
    'CD': cd.value_actual,
    'Vinyl': vinyl.value_actual,
    
})
```

```python
df
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(4, 5))       # size in inches

df.plot()

#fig._subtitle("Fall of Vinyl", fontsize=20)

fig.suptitle('Fall of Vinyl', fontsize=20)
```
