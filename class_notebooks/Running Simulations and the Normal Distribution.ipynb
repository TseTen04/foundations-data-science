```python
import pandas
import random
```

```python

def roll_dice(iterations=1):
    """Roll a number of dice equal to iterations and return the total result."""

    # We create a variable to hold our final total
    total = 0
    
    # Roll a die a number of times equal to iterations
    # After every roll, add the result to total
    for iteration in range(iterations):
        total += random.randint(1, 6) 

    return total


def run_simulation(iterations=500):
    """Run the roll_dice function a number of times equal to iterations. Add the result to a list and return it. """
    out = []

    for iteration in range(iterations):
        out.append(roll_dice(3))

    return pandas.Series(out)



```

```python
s=run_simulation(1000000)
```

```python
s.value_counts().sort_index().plot()
s.plot(kind='hist')
s.to_csv('dice_rolls.csv')
```

Let's run or simulation 500 times.

```python
results = run_simulation(iterations=500)


results
```

A histogram lets us see the frequency of certain event outcomes. The event outcomes are lumped into groups, called "bins."

```python
results.plot(kind='hist')
```

We can also plot the distribution as a line, which gives us a sense of which outcomes are most likely and unlikely. When we roll 10 dice, it looks like the result cluster around 35.

```python
results.plot(kind='line')
```

Some of our results are more frequent, and some are infrequent. Let's see if we can graph that more nicely.

Let's first try to plot the count of each outcome.

```python
results.value_counts().plot(kind='line')
```

That's kind of a mess, because the index (representing the roll result) isn't sorted. Let's sort the index first:

```python
results.value_counts().sort_index().plot(kind='line')
```

OK, the shows the distrubtion of outcomes, but it looks jagged. THat's because we only ran 500 iterations of our simulation, so there's still a lot of randomness in the presentation of the outcomes.

Let's run out simulation a million times, instead of 500

```python
results_large_sample_size = run_simulation(iterations=1000000)
```

```python
results_large_sample_size.value_counts().sort_index().plot(kind='line')
```

Here we have the outcomes (the total after rolling 10 dice) on the y axis, and the number of times the outcome occured on the y axis.
